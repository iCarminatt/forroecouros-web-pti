# Organizar imagens em assets folder
Notebook para mover arquivos JPG e WebP para uma pasta central `assets/images`.

```python
# Import Required Libraries
import os
from pathlib import Path
import shutil
```

```python
# Define Source and Destination Paths
root = Path('.')
dest_dir = root / 'assets' / 'images'
```

```python
# Search for Image Files
image_paths = list(root.rglob('*.jpg')) + list(root.rglob('*.jpeg')) + list(root.rglob('*.webp'))
print(f'Found {len(image_paths)} image files')
```

```python
# Create Assets Folder
os.makedirs(dest_dir, exist_ok=True)
print(f'Ensured destination directory exists: {dest_dir}')
```

```python
# Move Files to Assets Folder
for path in image_paths:
    try:
        target = dest_dir / path.name
        if target.exists():
            # handle naming conflict by appending number
            stem = path.stem
            suffix = path.suffix
            i = 1
            while target.exists():
                target = dest_dir / f"{stem}_{i}{suffix}"
                i += 1
        shutil.move(str(path), str(target))
        print(f"Moved {path} -> {target}")
    except Exception as e:
        print(f"Failed to move {path}: {e}")
```

```python
# Verify File Movement
before_count = len(image_paths)
after_paths = list(dest_dir.iterdir())
after_count = len(after_paths)
print(f"{before_count} files moved to assets; {after_count} now in assets folder.")
print("Assets folder contents:")
for p in after_paths:
    print(" -", p.name)
```
